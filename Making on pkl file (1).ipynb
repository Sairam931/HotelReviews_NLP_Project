```python
import pandas as pd
df=pd.read_csv('E:\\DS - ExcelR\\Project 2 - NLP\\Hotel_reviews_3_columns.csv', encoding = 'latin1')
```

```python
df
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sentiment_analyzer = SentimentIntensityAnalyzer()
```

```python
def get_polarity_scores(text):
    return sentiment_analyzer.polarity_scores(text)
```

```python
get_polarity_scores('that was a good movie but cast could be better')
```

```python
df['Pol_Scores'] = df['Reviews_new'].apply(lambda text: sentiment_analyzer.polarity_scores(text))
df['Compound_Score']  = df['Pol_Scores'].apply(lambda pol_score_dict: pol_score_dict['compound'])
df['Postive_Score']  = df['Pol_Scores'].apply(lambda pol_score_dict: pol_score_dict['pos'])
df['Neg_Score']  = df['Pol_Scores'].apply(lambda pol_score_dict: pol_score_dict['neg'])
df['Neutral_Score']  = df['Pol_Scores'].apply(lambda pol_score_dict: pol_score_dict['neu'])
df.head(5)
```

```python
df['Combined_Score'] = df['Rating'].apply(lambda c: 'pos' if c >3 else 'neg')
df.head(5)
```

```python
df['Combined_Score'].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer()
tfidf = TfidfVectorizer(max_features=3000)
```

```python
X = tfidf.fit_transform(df['Reviews_new']).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['target'] = encoder.fit_transform(df['Combined_Score'])
df['target'].value_counts()
df.head(5)
```

```python
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=2)
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_train,Y_train)
Y_pred1 = gnb.predict(X_test)
Y_pred2 = gnb.predict(X_train)
print(accuracy_score(Y_test,Y_pred1))
print(accuracy_score(Y_train,Y_pred2))
print(confusion_matrix(Y_test,Y_pred1))
print(precision_score(Y_test,Y_pred1))
```

```python
mnb.fit(X_train,Y_train)
Y_pred1 = mnb.predict(X_test)
Y_pred2 = mnb.predict(X_train)
print(accuracy_score(Y_test,Y_pred1))
print(accuracy_score(Y_train,Y_pred2))
print(confusion_matrix(Y_test,Y_pred1))
print(precision_score(Y_test,Y_pred1))
```

```python
bnb.fit(X_train,Y_train)
Y_pred1 = bnb.predict(X_test)
Y_pred2 = bnb.predict(X_train)
print(accuracy_score(Y_test,Y_pred1))
print(accuracy_score(Y_train,Y_pred2))
print(confusion_matrix(Y_test,Y_pred1))
print(precision_score(Y_test,Y_pred1))
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,Y_train)
LR.fit(X_train,Y_train)
Y_pred1 = LR.predict(X_test)
Y_pred2 = LR.predict(X_train)
print(accuracy_score(Y_test,Y_pred1))
print(accuracy_score(Y_train,Y_pred2))
print(confusion_matrix(Y_test,Y_pred1))
print(precision_score(Y_test,Y_pred1))
```
